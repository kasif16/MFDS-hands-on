```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
url='/content/train.csv'
df=pd.read_csv(url)
print(df)
mean_value=df['fnlwgt'].mean()
print(f"mean value of fnlwgt is {mean_value}")

mode_value=df['fnlwgt'].mode()
print(f"mode value of fnlwgt is{mode_value}")
meadian_value=df['fnlwgt'].median ()
print (f"meadian value of fnlwgt:-{meadian_value}")
var_value=df ['fnlwgt'].var()
print (f"variance value of fnlwgt:{var_value}")
sd_value=df['fnlwgt'].std()
print (f"standard deviation value:-{sd_value}")
skew_value=df['fnlwgt'].skew()
print(f"skewness value:-{skew_value}")
kur_value=df ['fnlwgt'].kurt ()
print (f"kurotosis value:-{kur_value}")
df['fnlwgt'].plot(kind='hist')
plt.xlabel('fnlwgt')
plt.ylabel('frequency')


```
